```python
people=['kiran','arun','vijay','varun']
age=[25,30,35,40,45]
height=[145,151,165,173]
weight=[45,55,65,75]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')
plt.bar(x=people,height=weight, width=0.3,
        color='yellow',edgecolor='k',alpha=1.0)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=60)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=13)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(8,5))
plt.hist(weight,color='orange',edgecolor='k', alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18,fontstyle="italic")
plt.xlabel("Weight (in kgs)",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
